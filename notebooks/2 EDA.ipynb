## Exploratory Data Analysis

Exploratory Data Analysis (EDA) is an essential step in the data analysis process. It involves investigating and summarizing the main characteristics, patterns, and relationships within a dataset. EDA aims to gain insights, discover patterns, and identify potential issues or outliers that may impact subsequent analysis or modeling.

Importance of Exploratory Data Analysis (EDA):

+ Data Understanding
+ Data Quality Assessment
+ Feature Selection
+ Insights and Hypothesis Generation

+ and lot more

```python
features = [
 'gender',
 'Partner',
 'Dependents',
 'PhoneService',
 'InternetService',
 'OnlineSecurity',
 'Contract',
 'PaymentMethod',
 'TotalCharges',
 'SeniorCitizen', 'tenure', 'MonthlyCharges']
```

```python
# defining the fuction to visualise our data, where we are dividing the dataset, into num and cat types and then visualising separatly.
def visualise(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            sns.countplot(x=col, data=data)
            plt.xlabel = col 
            plt.show()
        else:
            sns.histplot(x=col, data=data)
            plt.xlabel = col
            plt.title(f'The distribution of {col}')
            plt.show()
```

```python
visualise(data) # calling the fucntion.
```

```python
# Our main Goal is to check the customer is goint to churn or not, so lets focus on more on the churn and see the bi variate analyis.
# Creating a function, that gives visuaisations with respect to the churn (out put variable)
def visualise_wrt_churn(data, features):
    for col in features:
        sns.countplot(x = col , hue = 'Churn', data = data)
        plt.xlabel = col
        plt.title (f'visualization of {col} with respect to Churn')
        plt.show()
```

```python
visualise_wrt_churn(data, features )
```

Here are some of questions to understand the data, much better 

Question - 1

What is the distribution of churn in the dataset?

```python
churn_distribution = data['Churn'].value_counts()
plt.bar(churn_distribution.index, churn_distribution.values)
plt.ylabel('Count')
plt.title('Distribution of Churn')
plt.show()
```

Question - 2

How does the churn rate vary based on different customer demographics, such as gender, senior citizen status, partner, and dependents?

```python
data.head()
```

```python
import matplotlib.pyplot as plt

# Churn rate by gender
gender_churn_rate = data.groupby('gender')['Churn'].apply(lambda x: (x == 'Yes').mean())
plt.figure(figsize=(8, 6))
plt.bar(gender_churn_rate.index, gender_churn_rate.values, color=['#1f77b4', '#ff7f0e'])
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Gender')
plt.show()

# Churn rate by senior citizen status
senior_churn_rate = data.groupby('SeniorCitizen')['Churn'].apply(lambda x: (x == 'Yes').mean())
plt.figure(figsize=(8, 6))
plt.bar(senior_churn_rate.index, senior_churn_rate.values, color=['#1f77b4', '#ff7f0e'])
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Senior Citizen Status')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Churn rate by partner
partner_churn_rate = data.groupby('Partner')['Churn'].apply(lambda x: (x == 'Yes').mean())
plt.figure(figsize=(8, 6))
plt.bar(partner_churn_rate.index, partner_churn_rate.values, color=['#1f77b4', '#ff7f0e'])
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Partner')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Churn rate by dependents
dependents_churn_rate = data.groupby('Dependents')['Churn'].apply(lambda x: (x == 'Yes').mean())
plt.figure(figsize=(8, 6))
plt.bar(dependents_churn_rate.index, dependents_churn_rate.values, color=['#1f77b4', '#ff7f0e'])
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Dependents')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

```

Question - 3

What are the most common reasons for churn based on the available features like internet service, online security, contract type, and payment method?

```python
import matplotlib.pyplot as plt
reasons_for_churn = data[['InternetService', 'OnlineSecurity', 'Contract', 'PaymentMethod', 'Churn']].groupby(['InternetService', 'OnlineSecurity', 'Contract', 'PaymentMethod'])['Churn'].sum().nlargest(5)
reasons_for_churn.plot(kind='bar')
plt.title('Top Reasons for Churn')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Question - 4

How does the churn rate vary based on the type of internet service?

```python
churn_by_internet_service = data.groupby('InternetService')['Churn'].value_counts(normalize=True).unstack()
churn_by_internet_service.plot(kind='bar', stacked='True')
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Internet Service')
plt.legend(['No Churn', 'Churn'])
plt.show()
```
